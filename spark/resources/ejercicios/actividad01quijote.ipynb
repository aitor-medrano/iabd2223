1. A partir del fichero de *El Quijote*:

    1. Crear un RDD a partir del fichero
    2. ¿Cuantas veces aparece la palabra `Dulcinea`? ¿Y `Rocinante`?
    3. Devuelve una lista ordenada según el número de veces que sale cada palabra de
menos a más

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate() 
sc = spark.sparkContext
rdd = sc.textFile("el_quijote.txt")
```

```python
rddF = rdd.flatMap(lambda x: x.lower().split(" ")).map(lambda x: x.strip(" .,''!()[]-;:\""))
```

```python
rddF = rddF
```

```python
rddF.filter(lambda x: x == "dulcinea").count()
```

```python
rddF.filter(lambda x: x == "rocinante").count()
```

```python
rddTupla = rddF.map(lambda x: (x,1))
rddWC = rddTupla.reduceByKey(lambda a,b: int(a)+int(b)).sortBy(lambda x: x[1], False)
rddWC.collect()
```
