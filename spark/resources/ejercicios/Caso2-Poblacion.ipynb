```python
# A partir del fichero de clientes, para cada ciudad, calcular cuantos clientes tenemos
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate() 
sc = spark.sparkContext
```

```python
clientesRDD = sc.textFile("customer.txt")
```

```python
# Utilizamos map para generar tuplas (ciudad, 1) - La ciudad es el 6Âª elemento del CSV
```

```python
ciudadesRDD = clientesRDD.map(lambda x: (x.split(",")[6], 1))
```

```python
ciudadesRDD.collect()
```

```python
reducidoRDD = ciudadesRDD.reduceByKey(lambda a,b: a+b)
reducidoRDD.collect()
```

```python
# formateamos la salida a ciudad,cantidad
formateadoRDD = reducidoRDD.map(lambda x: str(x[0]) + "," + str(x[1]))
```

```python
formateadoRDD.saveAsTextFile("poblacionCiudades")
```
