```python
# A partir del fichero de ratings, calcular la nota media de cada pel√≠cula,
# y quedarse con la que la media sea superior a 3
# Salida: codigo pelicula, nota media
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate() 
sc = spark.sparkContext
```

```python
clientesRDD = sc.textFile("ratings.txt")
clientesRDD.top(3)
```

```python
# Utilizamos Map para crear tuplas (codigopelicula, (nota, 1))
```

```python
peliculasRDD = clientesRDD.map(lambda x: (x.split("::")[0], (int(x.split("::")[2]), 1)))
```

```python
peliculasRDD.collect()
```

```python
# Sumamos las notas y la cantidad
peliculasCalculos = peliculasRDD.reduceByKey(lambda a,b: (a[0]+b[0], a[1]+b[1]))
peliculasCalculos.collect()
```

```python
# Opcion A: calculamos la media con mapValues
peliculasMediasMV = peliculasCalculos.mapValues(lambda x: x[0]/x[1])
peliculasMediasMV.collect()
```

```python
# Opcion B: calculamos la media con map
peliculasMedias = peliculasCalculos.map(lambda x: (x[0], x[1][0]/x[1][1]))
peliculasMedias.collect()
```

```python
peliculasMedia3 = peliculasMedias.filter(lambda x: x[1] >= 3)
peliculasMedia3.collect()
```
