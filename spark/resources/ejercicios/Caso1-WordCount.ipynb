```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate() 
sc = spark.sparkContext
```

```python
clientesRDD = sc.textFile("customer.txt")
```

```python
# Utilizamos flatMap para separar los campos (con comas) en elementos
```

```python
palabrasRDD = clientesRDD.flatMap(lambda x: x.split(","))
```

```python
palabrasRDD.count()
```

```python
# quitamos las XXXXXXXXX
palabrasFiltradas = palabrasRDD.filter(lambda x: x != "XXXXXXXXX")
palabrasFiltradas.count()
```

```python
# creamos una tupla con (palabra,1)
tuplasPalabrasRDD = palabrasFiltradas.map(lambda x: (x, 1))
```

```python
# reducimos los resultados sumados las cantidades
wordcountRDD = tuplasPalabrasRDD.reduceByKey(lambda a,b: a+b)
wordcountRDD.collect()
```

```python
# ordenamos los resultados
ordenadoRDD = wordcountRDD.sortBy(lambda x: x[1], False)
ordenadoRDD.collect()
```

```python
ordenadoRDD.saveAsTextFile("wordcount")
```
